```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl

# Create a real data set 
np.random.seed(47)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Linear regression model
theta0, theta1 = 4, 3
X_new = np.array([[0], [2]])
y_predict = theta0 + theta1 * X_new

# Visualize the data set and Linear regression line
plt.plot(X, y, "b.")
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.show()
```

```python
# Cost function
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1 /(2 * m)) * np.sum(np.square(predictions - y))
    return cost 

# Add a column from the data set (X_0 = 1)
X_b = np.c_[np.ones((100, 1)), X]

# Select theta as [0, 0] at the beginning
theta = np.array([[0], [0]])

# Compute cost
cost = compute_cost(X_b, y, theta)
print("Başlangıç Maliyeti:", cost)
```
