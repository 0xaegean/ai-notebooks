```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# verileri oluştur
x = np.arange(-5, 5, 0.1)
y = 1/ (1 + np.exp(-x))
```

```python
# modeli eğit
w = np.random.rand(1)
b = np.random.rand()

for i in range(10000):
    y_pred = w * x + b
    loss = np.mean((y_pred - y)**2)
    w -= 0.001 * np.dot(x.T, (y_pred - y))
    b -= 0.001 * np.sum(y_pred - y)
```

```python
# tahminleri hesapal
y_pred = w * x + b
```

```python
plt.plot(x, y)
plt.plot(x, y_pred)
plt.show()
```
